# Part 2: Histograms with Iris

A histogram is a plot that lets you discover, and show, the underlying frequency distribution (shape) of a set of continuous data. This allows the inspection of the data for its underlying distribution. A built-in dataset that is often used when learning R is the iris dataset. The Iris flower data set or Fisher's Iris data set is a multivariate data set introduced by the British statistician and biologist Ronald Fisher 1936.

2.7 Now lets create a new variable called histogram:

```r
histogram <- ggplot(data=iris, aes(x=Petal.Width))
```

2.8 The bars in a histogram are called bins - set the bin width to 0.2 and plot with ggplot2:

```r
histogram + geom_histogram(binwidth=0.2, aes(fill=Species))
```

2.9 This should plot a nice looking histogram, now to add some details, adding black should put a border round the bins:

```r
histogram + geom_histogram(binwidth=0.2, color="black", aes(fill=Species))
```

2.10 Now add some labels to the plot:

```r
histogram + geom_histogram(binwidth=0.2, color="black", aes(fill=Species)) +
  xlab("Petal Width") + ylab("Frequency")  + ggtitle("Histogram of Petal Width")
```
